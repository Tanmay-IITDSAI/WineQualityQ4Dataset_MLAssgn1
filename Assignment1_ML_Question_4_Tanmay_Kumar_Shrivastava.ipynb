For the Wine Quality Data Set, convert all the values in the quality attribute to
0 (bad) if the value is less than or equal to 6 and to 1 (good) otherwise. Normalize
all the other attributes between 0 and 1 by min-max scaling. Mention why we use
min-max scaling.

```python
import pandas as pd
df = pd.read_csv('WineQT.csv')
desc = df.describe()
desc_styled = desc.style.set_table_styles(
    [{
        'selector': 'thead th',
        'props': [('background-color', '#f4f4f4'), ('color', '#333'), ('border', '1px solid black')]
    },
    {
        'selector': 'tbody td',
        'props': [('border', '1px solid black'), ('text-align', 'right'), ('padding', '8px')]
    }]
).set_caption("Descriptive Statistics of Wine Quality Data (Normalized)")

# Display the styled DataFrame
desc_styled
```

```python
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv('/content/WineQT.csv')

# Display the first few rows of the dataframe to understand its structure
print("Initial DataFrame:")
print(df.head())

# Step 3: Convert 'quality' attribute to binary classification
# If quality is <= 6, it will be 0 (bad), else it will be 1 (good)
df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)

# Display the distribution of the converted 'quality' attribute
print("\nConverted 'quality' attribute distribution:")
print(df['quality'].value_counts())

# Step 4: Normalize all other attributes between 0 and 1 using Min-Max Scaling
# Min-Max scaling formula: X_scaled = (X - X_min) / (X_max - X_min)
scaler = MinMaxScaler()

# Select the feature columns (excluding 'quality')
feature_columns = df.columns[df.columns != 'quality']

# Apply min-max scaling to the selected features
df[feature_columns] = scaler.fit_transform(df[feature_columns])

# Display the first few rows of the normalized dataframe
print("\nNormalized DataFrame:")
print(df.head())

# Step 5: Save the modified dataframe to a new CSV file
df.to_csv('WineQT_normalized.csv', index=False)

print("\nThe DataFrame has been normalized and saved to 'WineQT_normalized.csv'.")
```
